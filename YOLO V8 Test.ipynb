```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt") 
model.predict(source="https://ultralytics.com/images/bus.jpg", save=True)
```

```python
from PIL import Image

# 1. Run the prediction again and save the results to a variable
results = model.predict(source="https://ultralytics.com/images/bus.jpg")

# 2. Plot the results
# The [..., ::-1] part fixes the colors (swaps Blue/Red)
res_plotted = results[0].plot()[..., ::-1]

# 3. Display the image
display(Image.fromarray(res_plotted))
```
